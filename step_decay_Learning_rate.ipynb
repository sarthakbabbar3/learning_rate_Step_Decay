```python
pre_rnn_dropout = 0.35
rnn_dropout = 0.35
rnn_units = 64
epochs = 1000
lr = 0.7
momentum = 0.9
decay_rate = 0.9
decay_step = 40
cycle_step = 20
cycle_rate = 0.9

def step_decay(epoch, lr, rates):
    
    ## 0th epoch
    if not epoch:
        return lr
        
    if epoch % decay_step == 0:
        return lr * decay_rate
    
    if epoch > 50:
        ## cycle lr
        if (epoch-50) % cycle_step < cycle_step//2:
            print('dec', epoch, cycle_step//2, epoch % cycle_step)
            return lr * cycle_rate
        elif (epoch-50) % cycle_step > cycle_step//2:
            print('inc', epoch, cycle_step//2, epoch % cycle_step)
            return lr / cycle_rate

    return lr
```

```python
lr = 0.5
lrs = [lr]
for epoch in range(1000):
    lr = step_decay(epoch, lr, [])
    lrs.append(lr)
```
